```python
# Test Script for Priority and Description Fixes
# Run this in Jupyter or as Python script

import requests

# Your credentials
TODOIST_API_TOKEN = TODOIST_API_TOKEN
TODOIST_PROJECT_ID = TODOIST_PROJECT_ID

base_url = "https://api.todoist.com/api/v1"
headers = {"Authorization": f"Bearer {TODOIST_API_TOKEN}"}

# ============================================================
# PRIORITY CONVERSION TEST
# ============================================================

print("="*60)
print("PRIORITY CONVERSION TEST")
print("="*60)

# Todoist API priority scale (INVERTED!)
print("\nTodoist API Scale (what we receive):")
print("  1 = Urgent (red flag)")
print("  2 = High (orange)")
print("  3 = Medium (yellow)")
print("  4 = Normal (white)")

print("\nInternal/Notion Scale (what we want):")
print("  1 = Normal")
print("  2 = Medium")
print("  3 = High")
print("  4 = Urgent")

# Conversion function
def todoist_api_to_internal(api_priority):
    """Convert Todoist API priority to internal."""
    return 5 - api_priority

def internal_to_todoist_api(internal_priority):
    """Convert internal priority to Todoist API."""
    return 5 - internal_priority

print("\nConversion Table:")
print("API → Internal → API")
for api_p in [1, 2, 3, 4]:
    internal = todoist_api_to_internal(api_p)
    back_to_api = internal_to_todoist_api(internal)
    print(f"  {api_p} → {internal} → {back_to_api} {'✓' if back_to_api == api_p else '✗'}")

# ============================================================
# FETCH SAMPLE TASK AND CHECK PRIORITY
# ============================================================

print("\n" + "="*60)
print("FETCHING SAMPLE TASK")
print("="*60)

url = f"{base_url}/tasks"
params = {"limit": 1}

response = requests.get(url, headers=headers, params=params)
if response.status_code == 200:
    data = response.json()
    if data.get("results"):
        task = data["results"][0]
        
        api_priority = task.get('priority', 4)
        internal_priority = todoist_api_to_internal(api_priority)
        
        print(f"\nTask: {task['content']}")
        print(f"  API Priority: {api_priority}")
        print(f"  Should show in Notion as: {internal_priority}")
        print(f"  Description: {task.get('description', '(empty)')[:50]}")
    else:
        print("No tasks found")
else:
    print(f"Error: {response.status_code}")

# ============================================================
# CREATE TEST TASK WITH PRIORITY 4 (URGENT)
# ============================================================

print("\n" + "="*60)
print("CREATE TEST TASK")
print("="*60)

print("\nCreating task with internal priority 4 (urgent)...")
print("This should send API priority 1 to Todoist")

internal_priority = 4  # We want urgent
api_priority = internal_to_todoist_api(internal_priority)

test_task_data = {
    "content": "TEST: Priority 4 Urgent Task",
    "description": "This task has description from creation",
    "project_id": TODOIST_PROJECT_ID,
    "priority": api_priority  # Send converted priority
}

response = requests.post(
    f"{base_url}/tasks",
    headers=headers,
    json=test_task_data
)

if response.status_code == 200:
    created_task = response.json()
    print(f"✓ Task created: {created_task['id']}")
    print(f"  Content: {created_task['content']}")
    print(f"  API returned priority: {created_task.get('priority')}")
    print(f"  Should show in Notion as: {todoist_api_to_internal(created_task.get('priority', 4))}")
    print(f"  Description: {created_task.get('description', '(empty)')}")
    print("\n✓ Now run sync and check Notion shows priority 4")
else:
    print(f"✗ Failed: {response.status_code}")
    print(response.text)

# ============================================================
# DESCRIPTION CHECK
# ============================================================

print("\n" + "="*60)
print("DESCRIPTION SYNC CHECK")
print("="*60)

print("\nFetch all tasks and check descriptions:")

url = f"{base_url}/tasks"
all_tasks = []
cursor = None

while True:
    params = {"limit": 100}
    if cursor:
        params["cursor"] = cursor
    
    response = requests.get(url, headers=headers, params=params)
    if response.status_code != 200:
        break
    
    data = response.json()
    all_tasks.extend(data.get("results", []))
    
    cursor = data.get("next_cursor")
    if not cursor:
        break

# Filter by project
project_tasks = [t for t in all_tasks if t.get('project_id') == TODOIST_PROJECT_ID]

print(f"\nFound {len(project_tasks)} tasks in project")
print("\nTasks with descriptions:")

count_with_desc = 0
for task in project_tasks:
    desc = task.get('description', '')
    if desc:
        count_with_desc += 1
        print(f"  - {task['content'][:40]}")
        print(f"    Description: {desc[:60]}...")

print(f"\nTotal with descriptions: {count_with_desc}/{len(project_tasks)}")

# ============================================================
# SUMMARY
# ============================================================

print("\n" + "="*60)
print("TEST SUMMARY")
print("="*60)

print("\nPriority Conversion:")
print("  ✓ Conversion function defined")
print("  ✓ Test task created with priority")
print("  → Run sync and verify in Notion")

print("\nDescription Sync:")
print(f"  ✓ Found {count_with_desc} tasks with descriptions")
print("  → These should sync to Notion")

print("\nNext Steps:")
print("  1. Replace todoist_api.py with PRIORITY_FIXED version")
print("  2. Run: python main.py")
print("  3. Check Notion:")
print("     - TEST task should show priority 4")
print("     - Tasks with descriptions should show them")

print("\n" + "="*60)
```
